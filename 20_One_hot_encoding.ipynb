```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#plotting regression plot
import seaborn as sns
sns.regplot(x=df['R&D Spend'],y=y,color='k')
```

```python
sns.regplot(x=df['Administration'],y=y,color='c')
```

```python
sns.regplot(x=df['Marketing Spend'],y=y,color='m')
```

```python
#One hot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
one=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=one.fit_transform(x)
x                     #[NY,Florida,California,R&D,Administration,Marketing]
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#to find slope
print("Slope is: ",model.coef_)
```

```python
#to find the constant
print("Constant is: ",model.intercept_)
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error       #MAE
print("Error is: ",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error   #MAPE
print("Percentage error is: ",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error     #MSE
print("Squared error is: ",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error   #RMSE
data=mean_squared_error(y_test,y_pred)
data
```

```python
sq=np.sqrt(data)
sq
```

```python
from sklearn.metrics import r2_score
print("R_2 score is: ",r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[8,9,10,11]
plt.plot(x,y)
plt.show()
```

```python
#To get as coordinates without drawing a line
x=[5,2,3,4]
y=[8,9,10,11]
plt.plot(x,y,"*")
```

```python
import numpy as np
x=np.array([5,8,3,6])
y=np.array([4,5,7,9])
plt.plot(x,y)
```

```python
data=np.array([3,8,1,10,5,7])  #if only one axis is passed, machine reads it as y-axis
plt.plot(data)
```

```python
y=np.array([3,8,1,10,7,6,2]) #to get both points and lines
plt.plot(y,marker='o')   #we can use +,.,p ..etc as marker
```

```python
y=[1,5,2,7]
plt.plot(y,'*:g') #here * is the marker and : means dotted lines and g means green color
# :dotted line
# -solid line
# --dashed line
```

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='o',ms=20)    #ms means marker-size
```

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='o',ms=20,mec='r')   #mec means marker edge color
```

```python
y=np.array([3,8,1,10])
plt.plot(y,marker='o',ms=20,mfc='g',mec='r')    #mfc means marker face color
```

```python
plt.plot(y,linestyle='dotted')
```

```python
x=[150,155,156,157]   #assume height
y=[45,50,51,58]       #assume weight
plt.plot(x,y)
plt.xlabel('Height in cm')
plt.ylabel('Weight in kg')
plt.title('Height and weight graph')
```

```python
x=[150,155,156,157]   #assume height
y=[45,50,51,58]       #assume weight
plt.plot(x,y)
plt.xlabel('Height in cm')
plt.ylabel('Weight in kg')
plt.title('Height and weight graph',loc='left')   #title location to the left
```

```python
x=[150,155,156,157]   #assume height
y=[45,50,51,58]       #assume weight
plt.plot(x,y)
plt.xlabel('Height in cm',color='b',size=10)
plt.ylabel('Weight in kg',color='c',size=10)
plt.title('Height and weight graph',loc='left',color='m',size=20)
```

```python
x=[150,155,156,157]   #assume height
y=[45,50,51,58]       #assume weight
plt.plot(x,y)
dic1={'family':'serif','color':'b','size':20}
dic2={'family':'serif','color':'g','size':15}
plt.xlabel('Height in cm',fontdict=dic2)
plt.ylabel('Weight in kg',fontdict=dic2)
plt.title('Height and weight graph',fontdict=dic1)
```

```python
x=[23,54,63,10]
y=[20,12,34,32]
plt.plot(x,y)
dic1={'family':'serif','color':'b','size':20}
dic2={'family':'serif','color':'g','size':15}
plt.xlabel('VELOCITY',fontdict=dic2)
plt.ylabel('DISPLACEMENT',fontdict=dic2)
plt.title('VELOCITY-DISPLACEMENT GRAPH',fontdict=dic1)
```

```python
#GRID
import matplotlib.pyplot as plt
x=[150,155,160,165]
y=[45,50,55,60]
plt.plot(x,y)
plt.grid(color='green',linestyle='--',linewidth=1)
```

```python
#SUBPLOT
x1=[45,50,55,60]
y1=[12,13,14,15]
plt.subplot(1,2,1)  #plot in 1row & 2 cols ; 1st position
plt.plot(x1,y1)
plt.title('Graph 1')

x2=[70,75,80,85]
y2=[100,90,80,99]
plt.subplot(1,2,2)  #plot in 1 row & 2 cols; 2nd position
plt.plot(x2,y2)
plt.title('Graph 2')
```

```python
x1=[45,50,55,60]
y1=[12,13,14,15]
plt.subplot(2,1,1)   #plot in 2 rows& 1 col ; 1st position
plt.plot(x1,y1)
plt.title('Graph 1')

x2=[70,75,80,85]
y2=[100,90,80,99]
plt.subplot(2,1,2)    #plot in 2 rows & 1 col in 2nd position
plt.plot(x2,y2)
plt.title('Graph 2')
```

```python
x1=[45,50,55,60]
y1=[12,13,14,15]
plt.subplot(1,2,2)  #plot in 1row & 2 cols ; 2nd position
plt.plot(x1,y1)
plt.title('Graph 1')

x2=[70,75,80,85]
y2=[100,90,80,99]
plt.subplot(1,2,1)  #plot in 1 row & 2 cols; 1st position
plt.plot(x2,y2)
plt.title('Graph 2')
```

```python
x1=[45,50,55,60]
y1=[12,13,14,15]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(1,2,1)  #plot in 1row & 2 cols ; 1st position
plt.plot(x1,y1,'o--y')
plt.title('Graph 1')

x2=[70,75,80,85]
y2=[100,90,80,99]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.subplot(1,2,2)  #plot in 1 row & 2 cols; 2nd position
plt.plot(x2,y2,'*:g')
plt.title('Graph 2')
plt.suptitle('Graphs')
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Extracurricular Activities'] = lab.fit_transform(df['Extracurricular Activities'])
df.head()
```

```python
x=df.iloc[:,:-1].values
x
```

```python
x.ndim
```

```python
y=df.iloc[:,-1].values
y
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#Model creation
ann=tf.keras.models.Sequential()  #keras is the module in tf package
#input layer
```

```python
# first hidden layer
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
#second hidden layer
ann.add(tf.keras.layers.Dense(6,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=500)
```

```python
x=df.iloc[:,1:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
x.ndim
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
#To plot graph
plt.scatter(x_train,y_train,color='c')
plt.plot(x_test,y_pred,color='r')   #y=mx+c
```

```python
#to find slope
print("Slope is: ",model.coef_)
```

```python
#to find the constant
print("Constant is: ",model.intercept_)
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error       #MAE
print("Error is: ",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error   #MAPE
print("Percentage error is: ",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error     #MSE
print("Squared error is: ",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error   #RMSE
data=mean_squared_error(y_test,y_pred)
data
```

```python
sq=np.sqrt(data)
sq
```

```python
from sklearn.metrics import r2_score
print("R_2 score is: ",r2_score(y_test,y_pred))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/advertising.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
#regplot using seaborn; TV vs Sales
import seaborn as sns
sns.regplot(x=df['TV'],y=y,color='k')
```

```python
#Radio vs Sales
sns.regplot(x=df['Radio'],y=y,color='c')
```

```python
#Newspaper vs Sales
sns.regplot(x=df['Newspaper'],y=y,color='g')
```

```python
x.ndim
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[150,55,72]]))
```

```python
print("Slope is: ",model.coef_)
```

```python
#To print slopes separately:
print('Slope is: ')
list(zip(x,model.coef_))
```

```python
print('Constant is: ',model.intercept_)
```

```python
#To print dataframe
df1=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
df1
```

```python
from sklearn.metrics import mean_absolute_error       #MAE
print("Error is: ",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error        #MAPE
print("Mean absolute percentage error is:", mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error     #MSE
print("Squared error is: ",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error     #RMSE
data=mean_squared_error(y_test,y_pred)
data
```

```python
sq=np.sqrt(data)
sq
```

```python
from sklearn.metrics import r2_score       #R_2 Score
print("R_2 score is: ",r2_score(y_test,y_pred))
```

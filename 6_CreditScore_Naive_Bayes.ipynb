```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Credit Score Classification Dataset (1).csv')
df
```

```python
a=df['Income'].mean()
df['Income'].fillna(a,inplace=True)

b=df['Gender'].mode()[0]
df['Gender'].fillna(b,inplace=True)
print(df.isna().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Gender']=lab.fit_transform(df['Gender'])   #since it is not a list
df
```

```python
df['Education']=lab.fit_transform(df['Education'])
df
```

```python
df['Marital Status']=lab.fit_transform(df['Marital Status'])
df
```

```python
df['Home Ownership']=lab.fit_transform(df['Home Ownership'])
df
```

```python
x=df.iloc[:,:-1].values
x
```

```python
x.ndim
```

```python
y=df.iloc[:,-1].values
y
```

```python
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
x_test
```

```python
from sklearn.naive_bayes import GaussianNB
data=GaussianNB()
data.fit(x_train,y_train)
```

```python
y_pred=data.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
mat=confusion_matrix(y_test,y_pred)
mat
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
label=['High','Average','Low']
dis=ConfusionMatrixDisplay(mat,display_labels=label)

dis.plot()
```

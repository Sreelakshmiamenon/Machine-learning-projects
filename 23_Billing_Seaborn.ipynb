Three lines are the three percentiles, Q1,Q2,Q3;
Q1: 25 percentile
Q2: 50 percentile
Q3: 75 percentile
Outliers: The circles above maximum(wrong format data).

3,5,7,1,1,8,4,6

To calculate percentiles:
sort in ascending order: 1,1,3,4,5,6,7,8
calculate median: 4+5/2=4.5  i.e. Q2:50 percentile

1,1,3,4     5,6,7,8
Q1: 1+3/2=2  i.e. Q1:25 percentile
Q3: 6+7/2=6.5   i.e. Q3: 75 percentile

To calculate Quartile:
IQR(Inter Quartile Range)
Q3-Q1
6.5-2=4.5

To calculate maximum outliers:
Q3 + 1.5 IQR = 6.5+1.5*4.5 = 13.25

To calculate minimum outliers:
Q1 - 1.5 IQR = 2-1.5*4.5 = -4.75

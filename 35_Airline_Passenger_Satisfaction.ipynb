```python
df.columns
```

```python
#Checking the count of zeroes in Likert scale rating features
cols=['Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness']
for col in cols:
  count=df[col].value_counts()
  zeros=count.get(0)
  print(f"Number of zeros in '{col}':{zeros}")
```

```python
#Replacing all zero values with null values
modify = ['Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness']
for col in modify:
  for i in range(len(df)):
    if df.loc[i,col]==0:
      df.loc[i,col]=pd.NA
```

```python
for col in cols:
  counts=df[col].value_counts()
  zeros=counts.get(0)
  print(f"Number of zeros in '{col}':{zeros}")
```

### Handling of imbalance data

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(sampling_strategy='auto',random_state=42)
x_smote,y_smote=smote.fit_resample(x,y)
y_smote.value_counts()
```

### Splitting of Testing and Training data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote,test_size=0.30,random_state=42)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

### Modelling and Performance Evaluation

1. K-NEAREST NEIGHBOR

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
y_test = y_test[:len(y_pred_knn)]
print("Confusion matrix is",'\n'
,confusion_matrix(y_test,y_pred_knn))
print('Accuracy_score',accuracy_score(y_test,y_pred_knn))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_knn))
```

2. NAIVE BAYES CLASSIFIER


```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
nb.fit(x_train,y_train)
y_pred_nb=nb.predict(x_test)
y_test = y_test[:len(y_pred_nb)]
print("Confusion matrix is",'\n'
,confusion_matrix(y_test,y_pred_nb))
print('Accuracy_score',accuracy_score(y_test,y_pred_nb))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_nb))
```

3. SUPPORT VECTOR CLASSIFIER

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
y_pred_sv=sv.predict(x_test)
y_test = y_test[:len(y_pred_sv)]
print("Confusion matrix is",'\n'
,confusion_matrix(y_test,y_pred_sv))
print('Accuracy_score',accuracy_score(y_test,y_pred_sv))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_sv))
```

4. DECISION - TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy')
dt.fit(x_train,y_train)
y_pred_dt=dt.predict(x_test)
print('Confusion matrix is','\n'
      ,confusion_matrix(y_test,y_pred_dt))
print('Accuracy_score',accuracy_score(y_test,y_pred_dt))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_dt))
```

5. RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred_rf = rf.predict(x_test)
print(classification_report(y_test, y_pred_rf))
```

```python
print(confusion_matrix(y_test,y_pred_rf))
print('Accuracy_score',accuracy_score(y_test,y_pred_rf))
```
